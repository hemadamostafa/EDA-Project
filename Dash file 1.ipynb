```python
latitude = listing['latitude']
longitude = listing['longitude']
reviews = listing['review_scores_value']
# Define colormap
cmap = plt.cm.get_cmap('viridis')
scatter = plt.scatter(latitude, longitude, c=reviews, cmap=cmap,s=2)
plt.colorbar(scatter, label='Number of Reviews')
```

```python
reviews_features = ["review_scores_rating","review_scores_accuracy","review_scores_cleanliness","review_scores_cleanliness"]
for col in reviews_features:
    latitude = listing['latitude']
    longitude = listing['longitude']
    reviews = listing[col]
    # Define colormap
    cmap = plt.cm.get_cmap('viridis')
    scatter = plt.scatter(latitude, longitude, c=reviews, cmap=cmap,s=2)
    plt.colorbar(scatter, label='Number of Reviews')
    plt.show()
```
