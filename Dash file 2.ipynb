```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("file.csv")
```

```python
data["T/F"] = np.random.randint(0,2,len(data))
```

```python
_,axis = plt.subplots(nrows=1,ncols=2,figsize=(10,5))
sns.boxplot(x="Year",y="T/F",sharey=True,ax=axis[0],data=data)
```

```python
sns.catplot(x="Year",y="T/F",data=data,col="Year",kind="box")
```
